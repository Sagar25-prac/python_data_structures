```python
from collections import defaultdict

dd = defaultdict(list)

dd['key1'].append(1)

print(dd["key2"])


dd1 = defaultdict(lambda:"default value")
print(dd["key3"])
```

```python
from collections import Counter

votes =[
    'Python','java','Rust','Go','Python','Python','java','Rust','Rust','Rust',
    'Go'
]

vote_count = Counter(votes)

print(vote_count)
```

```python
from collections import namedtuple

pixel = namedtuple('Pixel',"R G B")

pixel = pixel(R=255,G=5,B=0)

print(pixel)

print(pixel._fields)

print(pixel.G)
print(pixel.R)
print(pixel.B)
```
