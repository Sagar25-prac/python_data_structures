```python
empty_sets = set()
sets = {1,2,"see"}

print(empty_sets)
print(sets)
```

```python
empty_sets = set()
sets = {1,2,"see"}

print(empty_sets)
print(sets)
```

```python
fruits = {"Apple","banana"}
print(fruits)

fruits.add("orange")
print(fruits)

fruits.remove("Apple")
print(fruits)

#dont give error if the value is not present
fruits.discard("cat")
print(fruits)

#we can empty the sets aswell

fruits.clear()
print(fruits)
```

```python

# Set operations subset

num = {1,2,3,4,5}
saw = {3,4,5,10}
seq = {7,8,9}

print(saw.issubset(num))
print(seq.issubset(num))
print(num.issuperset(saw))
print(num.isdisjoint(saw))
print(num.isdisjoint(seq))

print(saw.union(seq))

print(saw.intersection(num))

print(num.difference(saw))

print(num.symmetric_difference(saw))
```
