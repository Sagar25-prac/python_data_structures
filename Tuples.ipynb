```python
def circumference(radius):
  return 2*3.14*radius
radius = 5

round = circumference(radius)

print(f"Circumference is:{round:.2f}")
```

```python
reversed_tuple = tuple(reversed(multiple_elements))
print(reversed_tuple)
```

```python
slice_tuple = multiple_elements[::-1]
print(slice_tuple)
```

```python
def vol(len,hei,wid):
  return len*hei*wid

dimension = (10,2,4)
len,hei,wid = dimension

volume = vol(len,hei,wid)
print(f"volume is {volume}")
```

```python
volume1 =vol(*dimension)
print(volume1)
```

```python
def sum_numbers(*elements):
  print(elements)
  return sum(elements)

print(sum_numbers(1,2,3))
print(sum_numbers(20,15,44,85))
```

```python
def min_max(number):
  return min(number),max(number)

n  = [1,2,3,4,5,6]
min_result,max_result = min_max(n)
print(f"Minimum number is {min_result} and Maximum number is {max_result}")
```
